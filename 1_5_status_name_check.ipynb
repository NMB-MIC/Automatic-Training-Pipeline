**This Notebook is for status name checking**

```python
import pandas as pd
import numpy as np
from datetime import timedelta
from collections import deque
```

```python
DATA_PATH = "..." # <--- put file path here for example C:\DatasetTest\data1.csv or just file name if the dataset file is in the same folder as this notebook for example data1.csv

df_raw = pd.read_csv(DATA_PATH) # <--- change to match the type of file for example CSV file = pd.read_csv(DATA_PATH) or excel = pd.read_excel(DATA_PATH)
df_raw['occurred']   = pd.to_datetime(df_raw['occurred'])

# Normalize status text and basic sort
df_raw['mc_status'] = df_raw['mc_status'].astype(str).str.strip().str.lower()
df_raw = df_raw.sort_values(['mc_no','occurred']).reset_index(drop=True)
print("Raw shape:", df_raw.shape)
print("Statuses:", df_raw['mc_status'].unique())
print("Time range:", df_raw['occurred'].min(), "->", df_raw['occurred'].max())

# Status distribution
print("\nStatus counts:")
print(df_raw['mc_status'].value_counts())

print("\nHow dataset table looks:")
df_raw.head(5)
```
